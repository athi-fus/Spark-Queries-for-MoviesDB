```python
from pyspark.sql import SparkSession
from pyspark.sql import functions
from pyspark.sql.window import Window
from pyspark.sql.functions import col, row_number
from pyspark.sql.functions import split, regexp_replace
from pyspark.sql.functions import explode

from sparkmeasure import StageMetrics

#spark = SparkSession.builder.appName("DB II Project").getOrCreate()
spark = SparkSession.builder.appName('q9').config("spark.jars", "/home/athina/Downloads/spark-measure_2.12-0.17.jar").getOrCreate()
stagemetrics = StageMetrics(spark)


stagemetrics.begin()

#df_mv = (spark.read.format('csv').option('header', 'true').option('infereSchema', 'true').load("/home/administrator/Downloads/movielens/movie.csv"))

#df_mv = (spark.read.format('csv').option('header', 'true').option('infereSchema', 'true').load("/home/administrator/Downloads/movielens/movie.csv"))

#df_tag = spark.read.format('csv').option('header', 'true').option('infereSchema', 'true').load('/home/administrator/Downloads/movielens/tag.csv')
#df_rt = spark.read.format('csv').option('header', 'true').option('inferSchema', 'true').load('rating.csv')

df_rt = spark.read.format('csv').option('header', 'true').option('inferSchema', 'true').load('rating.csv')


#--Q9 START--------------------------------------------------------------------------------------------------------
#q9

split_col =functions.split(df_rt['timestamp'], ':')
df_rt_9 = df_rt.withColumn('Time', split_col.getItem(0))

#df_rt_9.show() 

dd = df_rt_9.groupBy('movieId', 'Time').count().orderBy('count', ascending = False)#.show()
dd = dd.filter(dd['count'] != 1).orderBy(dd['count'], ascending = False)#.show()
dd.groupBy().sum('count').show()


stagemetrics.end()

#print performance metrics
stagemetrics.print_report()
spark.stop()


#--Q9 END----------------------------------------------------------------------------------------------------------
```
